## 2. üìù Todo-Liste

```python
class TodoList:
    def __init__(self):
        self.todos = []
    
    def add(self, todo):
        self.todos.append(todo)
        print(f"‚úì '{todo}' hinzugef√ºgt")
    
    def show(self):
        if not self.todos:
            print("Keine Todos")
            return
        
        print("\n--- TODOS ---")
        for i, todo in enumerate(self.todos, 1):
            print(f"{i}. {todo}")
    
    def remove(self, index):
        if 0 <= index < len(self.todos):
            removed = self.todos.pop(index)
            print(f"‚úì '{removed}' entfernt")
        else:
            print("‚úó Ung√ºltiger Index")

# Testen
liste = TodoList()
liste.add("Python lernen")
liste.add("Projekt machen")
liste.show()
liste.remove(0)
liste.show()
```

## 3. üîê Passwort-Generator

```python
import random
import string

def generiere_passwort(laenge=12):
    zeichen = string.ascii_letters + string.digits + "!@#$%^&*()"
    passwort = ''.join(random.choice(zeichen) for _ in range(laenge))
    return passwort

# 5 Passw√∂rter generieren
for i in range(5):
    pw = generiere_passwort(16)
    print(f"Passwort {i+1}: {pw}")
```

## 4. üéÆ Zahlen-Raten

```python
import random

def zahlen_raten():
    ziel = random.randint(1, 100)
    versuche = 0
    max_versuche = 7
    
    print("üéÆ Zahlen-Raten-Spiel!")
    print(f"Ich denke an eine Zahl zwischen 1 und 100.")
    print(f"Du hast {max_versuche} Versuche!\n")
    
    while versuche < max_versuche:
        try:
            tipp = int(input(f"Versuch {versuche+1}/{max_versuche}: "))
            versuche += 1
            
            if tipp == ziel:
                print(f"üéâ Richtig! Du hast {versuche} Versuche gebraucht!")
                return
            elif tipp < ziel:
                print("‚¨ÜÔ∏è  Zu niedrig!")
            else:
                print("‚¨áÔ∏è  Zu hoch!")
        except ValueError:
            print("Bitte eine Zahl eingeben!")
    
    print(f"‚ùå Verloren! Die Zahl war {ziel}")

# Spiel starten (auskommentieren zum Testen)
# zahlen_raten()
```

## 5. üí∞ Einkaufsliste mit Preisen

```python
einkaufsliste = {
    "√Ñpfel": {"preis": 2.50, "anzahl": 3},
    "Bananen": {"preis": 1.80, "anzahl": 5},
    "Milch": {"preis": 1.20, "anzahl": 2},
    "Brot": {"preis": 2.00, "anzahl": 1}
}

# Rechnung ausgeben
print("üìã EINKAUFSLISTE")
print("=" * 50)

gesamt = 0
for artikel, daten in einkaufsliste.items():
    preis = daten["preis"]
    anzahl = daten["anzahl"]
    summe = preis * anzahl
    gesamt += summe
    
    print(f"{artikel:<15} {anzahl:>3}√ó {preis:>5.2f}‚Ç¨ = {summe:>7.2f}‚Ç¨")

print("=" * 50)
print(f"{'GESAMT':<15} {'':>3}  {'':>5}   {gesamt:>7.2f}‚Ç¨")
```

## 6. üìä Wort-Statistik

```python
def analysiere_text(text):
    # Bereinigen
    text = text.lower()
    
    # W√∂rter
    woerter = text.split()
    
    # Statistiken
    print("üìä TEXT-ANALYSE")
    print("=" * 50)
    print(f"Zeichen gesamt: {len(text)}")
    print(f"Zeichen (ohne Leerzeichen): {len(text.replace(' ', ''))}")
    print(f"W√∂rter: {len(woerter)}")
    print(f"Eindeutige W√∂rter: {len(set(woerter))}")
    print(f"Durchschnittliche Wortl√§nge: {sum(len(w) for w in woerter) / len(woerter):.1f}")
    print(f"L√§ngstes Wort: {max(woerter, key=len)}")
    
    # H√§ufigste W√∂rter
    wort_count = {}
    for wort in woerter:
        wort_count[wort] = wort_count.get(wort, 0) + 1
    
    print("\nTop 5 W√∂rter:")
    for wort, count in sorted(wort_count.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"  '{wort}': {count}√ó")

# Test
text = "Python ist toll Python macht Spa√ü Python ist einfach Python ist super"
analysiere_text(text)
```

## 7. ‚úÇÔ∏è Schere-Stein-Papier

```python
import random

def schere_stein_papier():
    optionen = ["schere", "stein", "papier"]
    
    print("‚úÇÔ∏èüìÑü™® Schere-Stein-Papier")
    
    spieler = input("W√§hle (schere/stein/papier): ").lower()
    computer = random.choice(optionen)
    
    print(f"\nDu: {spieler}")
    print(f"Computer: {computer}")
    
    if spieler == computer:
        print("\nü§ù Unentschieden!")
    elif (spieler == "schere" and computer == "papier") or \
         (spieler == "stein" and computer == "schere") or \
         (spieler == "papier" and computer == "stein"):
        print("\nüéâ Du gewinnst!")
    else:
        print("\nüò¢ Computer gewinnt!")

# Spiel starten (auskommentieren zum Testen)
# schere_stein_papier()
```
